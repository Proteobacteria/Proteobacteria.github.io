````python
import pandas as pd
import numpy as np
from scipy import stats

# colab의 출력을 소수점 이하 3자리로 제한
%precision 10
# pandas DataFrame의 소수점 출력을 3자릴 제한
pd.set_option('precision', 10)

````

````python
tree = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt', delimiter = "\t", skiprows=1)
tree
````

### "서울시에는 구별로 평균 약 400 그루 정도의 `이팝나무`가 있다"라는 가설에 대해 가설검정을 시행하세요.

- 단, `서울시`, `공원녹지사업소`, `시설관리공단` 부분의 데이터는 제외합니다.
- 가설검정 결과의 pvalue 값을 `pv1`에 저장하세요. (`.pvalue` 사용)

````python
tree_fringetree = pd.concat([tree['자치구'], tree['이팝나무']], axis=1)
drop_index = tree_fringetree[tree_fringetree['자치구'].isin(['서울시','공원녹지사업소','시설관리공단'])].index

tree1=tree_fringetree.drop(drop_index, axis=0)

tree1 = pd.to_numeric(tree1['이팝나무'].str.replace(',',''))

pv1 = stats.ttest_1samp(tree1, 400).pvalue
pv1
````

### 이번에는 10개의 구를 샘플링하여, 위에 대한 가설검정을 다시 한번 시행해보세요.

- 단, 아래의 코드를 참조하여 sample을 위한 seed 는 123로 설정하세요.
- 가설검정 결과의 pvalue값을 `pv2`에 저장하세요


````python
import numpy as np
np.random.seed(123)

random_index = np.random.choice(tree1.index, size = 10)  # "~"와 " "을 적절한 값으로 채우시면 됩니다
tree2=tree.iloc[random_index]

tree2 = pd.to_numeric(tree2['이팝나무'].str.replace(',',''))

pv2 = stats.ttest_1samp(tree2, 400).pvalue
pv2.round(5)
````

## 2. Two-Sample T-test 연습

### "서울시의 구별 평균 느티나무수와 왕벚나무의 수는 차이가 없다"라는 가설에 대해 가설검정을 시행하세요.

- 단, `서울시`, `공원녹지사업소`, `시설관리공단` 부분의 데이터는 제외합니다.
- 가설검정 결과의 pvalue를 `pv3`에 저장하세요

````python
trees = pd.concat([tree['자치구'], tree['느티나무'], tree['왕벚나무']], axis=1)
trees = trees.drop(drop_index, axis=0)

tree_zelkova = pd.concat([trees['자치구'], trees['느티나무']], axis=1)
tree_cherry = pd.concat([trees['자치구'], trees['왕벚나무']], axis=1)

tree_zelkova = pd.to_numeric(tree_zelkova['느티나무'].str.replace(',',''))
tree_cherry = pd.to_numeric(tree_cherry['왕벚나무'].str.replace(',',''))

pv3 = stats.ttest_ind(tree_zelkova, tree_cherry).pvalue
pv3
````


### Two-Sample One-tailed test를 사용하여 **평균 느티나무수가 왕벚나무 수보다 크다**라는 가설에 대한 가설검정을 시행하세요.
- 해당 가설검정의 결과를 `ans`에 저장하세요.

이 때 `ans`가 가질 수 있는 값은 아래와 같습니다 :
1. "느티나무" ( 느티나무의 수가 통계적으로 유의하게 많은 경우 )
2. "왕벚나무" ( 왕벚나무의 수가 통계적으로 유의하게 많은 경우 )
3. "없음" (통계적으로 유의하지 않은 경우)

Two-Sample One-tailed test를 scipy에서 하는 방법은 아래의 링크를 참고 하세요  
참고 링크 : https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

````python
# 아래의 코드를 실행해주세요
!pip install scipy --upgrade
````

# 가설의 대한 가설검정 ➡ p-value


*  ```p-value < 0.05``` ➡ **귀무가설 옳을 확률 5% 미만** - *대립가설 채택*

>  ⬛ ```p-value < 0.01``` ➡ **귀무가설이 옳을 확률 1% 미만** ➡ 깐깐한 기준

*  ```0.05 ≤ p-value ≤ 0.1``` ➡ 애매함

*  ```p-value > 0.1``` ➡ 귀무가설 채택



````python
from scipy import stats

result1 = stats.ttest_ind(tree_zelkova, tree_cherry, alternative='greater').pvalue

result2 = stats.ttest_ind(tree_zelkova, tree_cherry, alternative='less').pvalue

if result1 < 0.05:
  ans = '느티나무'
if result2 < 0.05:
  ans = '왕벚나무'
else:
  ans = '없음'


ans
````
